```python
import os
import sys
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import pandas as pd
```

```python
title = "CoV-SNN"
stats_dir_cont = "outputs/ContrastiveLoss_omicron_vs_other_Pmax_R0.2_D0.2_E10_LR_0.0001_B32_M2.0/stats"

# load
train = pd.read_csv(os.path.join(stats_dir_cont, "Train.csv"))
eval = pd.read_csv(os.path.join(stats_dir_cont, "Eval.csv"))
test = pd.read_csv(os.path.join(stats_dir_cont, "Test.csv"))

# plot
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# loss
axs[0].plot(train["epoch"]+1, train["loss"], label="Train")
axs[0].plot(eval["epoch"]+1, eval["loss"], label="Eval")
axs[0].plot(test["epoch"]+1, test["loss"], label="Test")
axs[0].set_title(title)
axs[0].set_xlabel("Epoch")
axs[0].set_ylabel("Loss")
axs[0].legend()
axs[0].set_xticks(train["epoch"]+1)

# acc
axs[1].plot(train["epoch"]+1, train["accuracy"], label="Train")
axs[1].plot(eval["epoch"]+1, eval["accuracy"], label="Eval")
axs[1].plot(test["epoch"]+1, test["accuracy"], label="Test")
axs[1].set_title(title)
axs[1].set_xlabel("Epoch")
axs[1].set_ylabel("Accuracy")
axs[1].legend()
axs[1].set_xticks(train["epoch"]+1)

font_size = 18
plt.rc('font', size=font_size)
plt.rc('axes', titlesize=font_size)
plt.rc('axes', labelsize=font_size)
plt.rc('xtick', labelsize=font_size-2)
plt.rc('ytick', labelsize=font_size-2)
plt.rc('legend', fontsize=font_size)
plt.rc('figure', titlesize=font_size)

plt.tight_layout()
plt.savefig("figures/loss_accuracy_cont.pdf")
plt.show()
```
